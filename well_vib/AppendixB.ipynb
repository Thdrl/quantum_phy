```python
##Résonnance Paramagnétique Electronique dans le rubis
import scipy
import numpy as np
from scipy . constants import hbar
from scipy . integrate import quad
import numpy.linalg as alg
import matplotlib.pyplot as plt
from scipy . optimize import curve_fit
from scipy import special


D=5.73e9 #5.73GHz
e=1.9e-19
g=2
a =1 #On choisit de prendre a=1, pour faire les tracés
v0=300
m =9e-31
om =1

def H(teta,B0):
    '''On définit une fonction qui prend en argument l'angle considéré ainsi que la valeur du champ et qui digaonalise l'hamiltonien             ainsi calculé.'''
    teta=teta*np.pi/180
    A=(-2*np.pi*D*hbar)-3*g*e*B0*np.cos(teta)*hbar/(4*m)
    F=2*np.pi*D*hbar-(g*e*B0*np.cos(teta)*hbar)/(4*m)
    C=2*np.pi*D*hbar+(g*e*B0*np.cos(teta)*hbar)/(4*m)
    E=(-2*np.pi*D*hbar)+3*g*e*B0*np.cos(teta)*hbar/(4*m)

    H= np.array([[A, g*e*B0*np.sqrt(3)*np.sin(teta)*hbar/(4*m), 0, 0],
                [g*e*B0*np.sqrt(3)*np.sin(teta)*hbar/(4*m), F, hbar*np.sin(teta)*e*g*B0/(2*m), 0],
                [0,  hbar*np.sin(teta)*e*g*B0/(2*m), C,  g*e*B0*np.sqrt(3)*np.sin(teta)*hbar/(4*m)],
                [0, 0, g*e*B0*np.sqrt(3)*np.sin(teta)*hbar/(4*m), E]])

    [spectre,vecs_p]= alg.eig (H)
    spectre=spectre/(2*np.pi*hbar)
    ordre=spectre.argsort ()
    vecs_p=vecs_p[:,ordre]
    spectre=spectre[ordre]


    return(spectre,vecs_p)


#On choisit l'état propre que l'on souhaite avoir



Bchamp=np.linspace(0,0.6,10000)

for i in range(4):
    numero=i
    Efreq=[H(50,b)[0][numero] for b in Bchamp]




    plt.figure(10+i)
    plt.plot(Bchamp,Efreq,'k:')
    plt.show()



import numpy as np



##On veut visualiser les transitions

n_p=50
 
def dE_list(theta,n):#n entre 0 et 2
    B=np.linspace(0,0.6,n_p)
    dE_vec=[]
    for B_item in B:
        energies=H(B_item,theta)[0]
        dE_vec.append(energies[n+1]-energies[n])
    return np.array(dE_vec)/(2*np.pi*hbar)
 
def dE2_list(theta,n):#n entre 0 et 1
    B=np.linspace(0,0.6,n_p)
    dE_vec=[]
    for B_item in B:
        energies=H(B_item,theta)[0]
        dE_vec.append(energies[n+2]-energies[n])
    return np.array(dE_vec)/(2*np.pi*hbar)
 
def dE3_list(theta):#n=0
    B=np.linspace(0,0.6,n_p)
    dE_vec=[]
    for B_item in B:
        energies=H(B_item,theta)[0]
        dE_vec.append(energies[3]-energies[0])
    return np.array(dE_vec)/(2*np.pi*hbar)
 
def transition_list(theta):
    B=np.linspace(0,0.6,n_p)
    fr=9.19E9
    trans=[] 
    for n in range(3): #transition entre niveaux adjacents
 
        dE=dE_list(theta,n)
        for i in range(n_p-1):
            if (dE[i]>fr and dE[i+1]<fr) or (dE[i]<fr and dE[i+1]>fr):
                trans.append(B[i])
 
    for n in range(2): #transition 0-2,1-3
        dE2=dE2_list(theta,n)
        for i in range(n_p-1):
            if (dE2[i]>fr and dE2[i+1]<fr) or (dE2[i]<fr and dE2[i+1]>fr):
                trans.append(B[i])
 
    dE3=dE3_list(theta) #transition 0-3
    for i in range(n_p-1):
        if (dE3[i]>fr and dE3[i+1]<fr) or (dE3[i]<fr and dE3[i+1]>fr):
            trans.append(B[i])
    return trans
 
plt.figure(figsize=[15,10])
theta_vec=np.linspace(0,np.pi,n_p)
for theta in theta_vec:
    for trans in transition_list(theta):
        plt.scatter(theta,trans,marker=".",color='k')
plt.vlines(50*np.pi/180,0,0.6,colors='b',linestyles=(0,(5,15)))
plt.ylabel('B (T)')
plt.xlabel('theta (rad)')
plt.show()
```
